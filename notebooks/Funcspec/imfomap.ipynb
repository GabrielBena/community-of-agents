```python
def get_all_links(community):

    in_links = []
    rec_links = []
    conns_links = []

    for i, ag in enumerate(community.agents):
        ws_in = ag.cell.weight_ih_l0
        ws_rec = ag.cell.weight_hh_l0

        n_neurons = ws_rec.shape[0]

        conns_out = community.connections[f"{i}{1-i}"].w

        for n_in, w_in in enumerate(ws_in.T):
            for n_unit, w in enumerate(w_in):
                in_links.append(((0, n_in), (1, n_unit + i * n_neurons), w.data.item()))

        
        for n_unit_1, w_in in enumerate(ws_rec.T):
            for n_unit_2, w in enumerate(w_in):
                rec_links.append(
                    (
                        (1, n_unit_1 + i * n_neurons),
                        (1, n_unit_2 + i * n_neurons),
                        w.data.item(),
                    )
                )

        
        for n_unit_1, w_in in enumerate(conns_out.T):
            for n_unit_2, w in enumerate(w_in):
                conns_links.append(
                    (
                        (1, n_unit_1 + i * n_neurons),
                        (1, n_unit_2 + (1 - i) * n_neurons),
                        w.data.item(),
                    )
                )

    return in_links, rec_links, conns_links
```

```python
community_links = get_all_links(community)
```

```python
from infomap import Infomap
im = Infomap("--directed --two-level")

for links in community_links : 
    im.add_multilayer_links(links)
```

```python
im.write_flow_tree('flow_tree')
```

```python
# Run the Infomap search algorithm to find optimal modules
im.run()

print(f"Found {im.num_top_modules} modules with codelength: {im.codelength}")

print("Result")
print("\n#node module")
for node in im.tree:
    if node.is_leaf:
        print(node.node_id, node.module_id)
```
